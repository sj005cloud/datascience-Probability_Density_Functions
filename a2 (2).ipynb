```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data.csv', encoding='latin1')
df.head()

```

```python
df.columns
```

```python
x = df['no2'].values.astype(float)


```

```python
x = x[~np.isnan(x)]
```

```python
r = 102317173
a = 0.05 * (r % 7)
b = 0.3 * ((r % 5) + 1)
a, b
```

```python
bx = np.clip(b*x, -1, 1)
z = x + a*np.arcsin(bx)
```

```python
mu = np.mean(z)
sigma_squared = np.var(z, ddof=0)
lam = 1 / (2 * sigma_squared)
c = np.sqrt(lam / np.pi)

mu, lam, c
```

```python
print("Estimated Parameters:")
print(f"μ (mu)     = {mu}")
print(f"λ (lambda) = {lam}")
print(f"c          = {c}")
```
